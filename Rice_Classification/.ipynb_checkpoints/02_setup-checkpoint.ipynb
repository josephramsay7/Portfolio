```python
import pandas as pd
import numpy as np
from pycaret.classification import ClassificationExperiment
from pycaret.classification import setup, models, create_model, tune_model, finalize_model, save_model, create_model, evaluate_model, save_model
```

```python
rice = pd.read_csv('./data/rice_tt.csv')
unseen_rice = pd.read_csv('./data/unseen_rice.csv')
rice.shape, unseen_rice.shape
```

```python
s = ClassificationExperiment()
```

```python
s.setup(rice, target='Class', session_id=42)
```

```python
%%time
best = s.compare_models()
```

```python
best
```

```python
ridge = s.create_model('ridge')
ridge
```

```python
final_ridge = s.finalize_model(ridge)
final_ridge
```

```python
s.evaluate_model(final_ridge)
```

```python
s.plot_model(best, plot='confusion_matrix')
```

```python
s.plot_model(best, plot='feature')
```
